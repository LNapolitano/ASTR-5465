```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
M_sig=[0.1,0.3,0.5,1.0] #values of different M uncertainties
#M_sig=[0.1] #testing simplicity

for i in M_sig:
    
    f = plt.figure(figsize=(7,10.5))
    
    M=np.random.normal(scale=i,size=10000) #generate normal distributed M vals
    
    dist=1000*np.random.rand(10000) #currently starting at 0 pc due to some sizing issues
    
    plt.subplot(4,2,1)
    plt.hist(M) #histogram of M
    plt.xlabel("Absolute Magnitude (M)")
    plt.ylabel("N")
    #plt.show()
    
    max_d=int(max(dist))#Setting up bins of 1pc
    N=np.zeros(1+max_d) #for N in each bin

    for j in dist:
        N[int(j)]+=1 #fill N according to distance distribtion
        
    
    #plt.scatter((np.arange(1+max_d)),N)
    #plt.xlabel("Distance (pc)")
    #plt.ylabel("N")
    #plt.show()
    
    plt.subplot(4,2,2)
    plt.scatter(np.log10(np.arange(1+max_d)),np.log10(N))
    plt.xlabel("Log(Distance (pc))")
    plt.ylabel("Log(N)")
    #plt.show()
    
    m=M+5*np.log10(dist/10) #compute apparent mags
    
    plt.subplot(4,2,3)
    plt.scatter(dist,m)
    plt.xlabel("Distance (pc)")
    plt.ylabel("Apparent magntiude (m)") #APPARENT MAGS ARE TOO BRIGHT, ADJUST WITH DISTANCE START POINT
    
    m_cut=[]; dist_cut=[]; M_cut=[]
    
    cut_val=8 #apparent magntiude cut value, CRUCIAL
    
    for (i,j,k) in zip(m,dist,M):
        if i<cut_val:
            m_cut.append(i)
            dist_cut.append(j)
            M_cut.append(k)
        
    plt.scatter(dist_cut,m_cut)
    #plt.show() #Shows cut in different color
    
    plt.subplot(4,2,4)
    plt.scatter(dist,M)
    plt.scatter(dist_cut,M_cut)
    plt.xlabel("Distance (pc)")
    plt.ylabel("Absolute magntiude (m)")
    #plt.show()
    
    binsize=5 #CRUCIAL
    bins=np.arange(0,max_d,binsize) #array of bin values for digitize
    
    pre_inds=(np.digitize(dist,bins),M)   #digitize, splitting values into bins
    post_inds=(np.digitize(dist_cut,bins),M_cut) 
    
    
    pre_avgs=[]
    post_avgs=[]
    
    #HERE BE BAD PYTHON
    for i in range(len(bins)):
        pre_vals=[]; post_vals=[]
        
        for j in range(len(pre_inds[0])): 
            if(i==pre_inds[0][j]):
                pre_vals.append(pre_inds[1][j])
                
        for k in range(len(post_inds[0])):
            if(i==post_inds[0][k]):
                post_vals.append(post_inds[1][k])
        
        pre_avgs.append(np.average(pre_vals))
        post_avgs.append(np.average(post_vals))
    
    pre_avgs=np.array(pre_avgs)
    post_avgs=np.array(post_avgs)
        
    #print((post_avgs))
    #print(len(pre_avgs))
    
    plt.subplot(4,2,5)
    plt.scatter(np.arange(0,1000,binsize),post_avgs-pre_avgs)
    plt.xlabel('Distance (pc)')
    plt.ylabel('<M>sample-<M>T')
    #plt.show()
    
        
    #Mest=a*P+b??
    #Plot M-<M> and M-Mest as fn of dist
    #average in their bin I presume.
    avg_diff=[] #M-<M>
    for i,j in zip(M_cut,dist_cut):
        avg_diff.append(i-post_avgs[int(j/5)])
        
    plt.subplot(4,2,6)    
    plt.scatter(dist_cut, M_cut)
    plt.xlabel('Distance (pc)')
    plt.ylabel('M')
    #plt.show()
    
    plt.subplot(4,2,7)
    plt.scatter(dist_cut, avg_diff)
    plt.xlabel('Distance (pc)')
    plt.ylabel('M-<M>')
    #plt.show()
    
    M_ests=np.random.normal(loc=avg_diff,scale=0.1) #Mest as estimator of M-<M>
    
    plt.subplot(4,2,8)
    plt.scatter(dist_cut, M_ests)
    plt.xlabel('Distance (pc)')
    plt.ylabel('M-Mest')
    #plt.show()
    
    plt.tight_layout()
    f.savefig(str(i)+".png", dpi=600)
    

    
    
```
