```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = plt.figure(figsize=(7,10.5))
#exponential profile (Disk)
def exp(I_o,R,h):
    return(I_o*np.exp(-R/h))
#deVaucouleurs Profile (Bulge/Ellipticals)
def r_4(I_o,R,k):
    return(I_o*np.exp(-k*R**(1/4)))

x=np.arange(0,1000,1) #radius values

h=np.arange(100,1000,100) #parameter h values
k=np.arange(0.1,1,0.1)    #parameter k values

plt.subplot(2,1,1)
for i in h:
    plt.plot(exp(1000,x,i))
plt.xlabel('Radius (pc)')
plt.ylabel('Luminosity ($L_{☉}$)')

plt.subplot(2,1,2)
for j in k:
    plt.plot(r_4(1000,x,j)) 
plt.xlabel('Radius (pc)')
plt.ylabel('Luminosity ($L_{☉}$)')
f.savefig("HW3plt.png", dpi=600)
```
